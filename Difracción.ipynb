```python
from math import *
from matplotlib import *
from pylab import *
l = float(input('Ingrese la longitud de onda \u03BB (en nm): '))
a = float(input('Ingrese el ancho de ranura a (en nm): ' ))
theta = arange(-pi, pi, 0.02)
I = (sin(pi*a/l*sin(theta)))**2/(pi*a/l*sin(theta))**2
y = pi*a/l*sin(theta)
plot(y, I)
xlabel('$\pi$(a/$\u03BB)sen(\u03B8$)')
ylabel('I / I_máx')
title('Difracción de una Rendija ')
plt.ylim([-0.1,1.1])
```
