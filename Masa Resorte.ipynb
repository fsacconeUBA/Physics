```python
from pylab import *

# Initialize
k = float(input("Ingrese la constante del resorte (N/m): ")) # Newtons por metro
m = float(input("Ingrese la masa (en kg): ")) # Kilogramos
g = 9.8 # metros por segundo
D = float(input("Coeficiente de arrastre D (en 1/m): ")) 
v0 = 0.0 # en m/s
time = 30.0 # s
vt = 0.2 # m/s
kv = D*vt

# Numerical setup
dt = 0.0001 # s
n = int(round(time/dt))
t = zeros(n,float)
y = zeros(n,float)
v = zeros(n,float)

# Valores Iniciales
y[0] = 0.0
v[0] = v0

# Loop de Simulación
for i in range(n-1):
	if (v[i]<vt):
		FD = -kv*v[i]
	else:
		FD = -D*v[i]*abs(v[i])
	F = -k*y[i] - m*g + FD
	a = F/m
	v[i+1] = v[i] + a*dt
	y[i+1] = y[i] + v[i+1]*dt
	t[i+1] = t[i] + dt

y = y + max(abs(y))/2

#Gráfico
fig, ax = plt.subplots()

ax.plot(t, y, '-r', label='altura de la masa')
ax.plot(t, v, '-b', label='velocidad de la masa')
plt.title('Sistema Masa-Resorte ')
plt.xlabel('tiempo (s)', color='#1C2833')
plt.ylabel('y(t), v(t)', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
